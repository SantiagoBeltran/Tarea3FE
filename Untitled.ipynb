```python
m=np.arange(0.1,5,0.01)
```

```python
def S_cl(m):
    return np.log(m)+1
def S_q(m):
    return (m+1/2)*np.log(m+1/2)-(m-1/2)*np.log(m-1/2)
```

```python
plt.figure()
plt.plot(m,S_cl(m), label="Clásico")
plt.plot(m,S_q(m), label="Cuántico")
plt.title("Entropía s")
plt.ylabel("S en unidades de $k_B$")
plt.xlabel("$\epsilon$ en unidades de $\hbar \omega$")
plt.legend()
plt.savefig("entropiaPunto3.png")
```

```python
x=np.arange(0.2,3,0.1)
def e_cl(x):
    return x
def e_q(x):
    return 0.5+1/(np.exp(1/x)-1)
```

```python
plt.figure()
plt.plot(x,e_cl(x), label="Clásico")
plt.plot(x,e_q(x), label="Cuántico")
plt.legend()
plt.title("Energía $\epsilon$")
plt.xlabel("T en unidades de $k_B/(\hbar \omega)$")
plt.ylabel("$\epsilon$ en unidades de $\hbar \omega$")
plt.savefig("energiaPunto3.png")
```
